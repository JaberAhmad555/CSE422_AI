```python
import random as rd
import math
```

```python
#task1
#name,w,h
component=[("ALU",5,5),("Cache",7,4),("Control Unit",4,4),("Register File",6,6),("Decoder",5,3),("Floating Unit",5,5)]
interconn=[(3, 0),(2, 0),(0, 1), (3, 5), (1, 4), (4, 5)] #R->A,C->A,A->CA......
alpha=1000 #for overlap
beta=2 #for wire
gamma=1 # for bound area

def rand_info():
  info=[]
  for i,w,h in component:
    x=rd.randint(0,25-w)
    y=rd.randint(0,25-h)
    info.append((x,y,w,h))
  return info
def overlap_checker(A,B):
  if A[0]+A[2]<=B[0] or A[1]+A[3]<=B[1] or B[0]+B[2]<=A[0] or B[1]+B[3]<=A[1]: #checking je ekta arektar upor uthe naki
    return False
  else:
    return True

def fitness_function(info): #joto kom toto bhalo
  c=0
  for i in range(len(info)):
    for j in range (i+1,len(info)):
      if overlap_checker(info[i],info[j]):
        c=c+1
  wiring=0
  for l,m in interconn:
    if l<len(info) and m<len(info):

      x1=info[l][0]+info[l][2]/2
      x2=info[m][0]+info[m][2]/2
      y1=info[l][1]+info[l][3]/2
      y2=info[m][1]+info[m][3]/2
      wiring=wiring+math.sqrt((x2-x1)**2+(y2-y1)**2)

  minx=float('inf')
  miny=float('inf')
  maxx=-float('inf')
  maxy=-float('inf')
  for k in info:
    x,y,w,h=k
    if x < minx:
      minx=x
    if x+w > maxx:
      maxx=x+w
    if y < miny:
      miny=y
    if y+h > maxy:
      maxy=y+h
  f=alpha*c+beta*wiring+gamma*((maxx-minx)*(maxy-miny))
  return f
def selection (population,fitness,size=3):
  selected=[]
  for i in range(size):
    idx=rd.randint(0,len(population)-1)
    selected.append((population[idx],fitness[idx]))
  min_fit=selected[0][1]
  min_sol=selected[0][0]
  for j,k in selected[1:]:
    if k<min_fit:
      min_fit=k
      min_sol=j
  return min_sol
def crossover(p1,p2):
  point=rd.randint(1,len(p1)-1)
  c1=p1[:point]+p2[point:]
  c2=p2[:point]+p1[point:]
  return c1,c2
def mutate(soln,m_rate=0.07):
  m_done=[]
  for i in range(len(soln)):
    if rd.random()<m_rate:
      name,w,h=component[i]
      m_done.append((rd.randint(0,25-w),rd.randint(0,25-h),w,h))
    else:
      m_done.append(soln[i])
  return m_done




def main_algo(popusize=6,gen=15):
  #making of population list
  population=[]
  for i in range(popusize):
    population.append(rand_info())
  for g in range(gen):
    fitness=[]
    for p in population:
      fitness.append(fitness_function(p))
    eliteidx=[]
    for e in range(len(fitness)):
      if len(eliteidx)<2:
        eliteidx.append(e)
      else:
        if fitness[e]<fitness[eliteidx[0]]:
          eliteidx.insert(0,e)
          eliteidx.pop()
        elif fitness[e]<fitness[eliteidx[1]]:
          eliteidx[1]=e
    new_pop=[]
    for idx in eliteidx:
      new_pop.append(population[idx])
    while len(new_pop)<popusize:
      p1=selection(population,fitness)
      p2=selection(population,fitness)
      c1,c2=crossover(p1,p2)
      new_pop.append(mutate(c1))
      if len(new_pop)<popusize:
        new_pop.append(mutate(c2))
    population=new_pop
    bestone=min(fitness)
    print(f"Generation:{g+1},Best fitness:{bestone:.2f}")

  best_idx=0
  best_fitness=fitness_function(population[0])
  for i in range(1,len(population)):
    curr=fitness_function(population[i])
    if curr<best_fitness:
      best_fitness=curr
      best_idx=i
  return population[best_idx]



def print_sol(info):
  for i in range(len(info)):
    name,w,h=component[i]
    x,y,z,a=info[i]
    print(f"{name}:({x},{y})to({x+z},{y+a})")
  c=0
  for i in range(len(info)):
    for j in range (i+1,len(info)):
      if overlap_checker(info[i],info[j]):
        c=c+1
  wiring=0
  for l,m in interconn:
    if l<len(info) and m<len(info):
      x1=info[l][0]+info[l][2]/2
      x2=info[m][0]+info[m][2]/2
      y1=info[l][1]+info[l][3]/2
      y2=info[m][1]+info[m][3]/2
      wiring=wiring+math.sqrt((x2-x1)**2+(y2-y1)**2)

  minx=float('inf')
  miny=float('inf')
  maxx=-float('inf')
  maxy=-float('inf')
  for k in info:
    x=k[0]
    y=k[1]
    w=k[2]
    h=k[3]
    if x<minx:
      minx=x
    if x+w>maxx:
      maxx=x+w
    if y<miny:
      miny=y
    if y+h>maxy:
      maxy=y+h
  area=(maxx-minx)*(maxy-miny)

  print(f"overlaps:{c}")
  print(f"wiring:{wiring}")
  print(f"bounded area:{area}")
  print(f"fitness:{fitness_function(info):.2f}")

if __name__=="__main__":
  best=main_algo()
  print_sol(best)
```
