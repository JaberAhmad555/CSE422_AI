```python
import numpy as np
import heapq #priority queue
```

```python
#1
def manhattan(X,Y):
  return abs(X[0]-Y[0])-abs(X[1]-Y[1])
with open('input1.txt','w') as f:
  f.write("""7 8
0 3
6 5
###0####
#000000#
###0#00#
#000#00#
#0######
#000000#
#####0##
  """)
def astar(maze,start,goal):
  n=len(maze) #7
  m=len(maze[0]) #8
  g=np.full((n,m),float('inf')) #cost to get from strt to any cell is being set to infinity as we dont know
  g[start[0],start[1]]=0  #g[0,3]=0,strt to srt dist=0
  source={}
  dir= [(-1, 0, 'U'), (1, 0, 'D'), (0, -1, 'L'), (0, 1, 'R')]
  pq=[[manhattan(start,goal),start[0],start[1]]] #f,x,y
  while pq:
    f,x,y=heapq.heappop(pq) #first item ta ber kre dei from pq
    if (x,y)==goal:
      break
    else:
      for ox,oy,rasta in dir: #checks 4 neighbours of 0,3
        nx,ny=x+ox,y+oy  #exp 0-1,3+0=-1,3
        if 0<=nx<n and 0<=ny<m and maze[nx][ny]=='0':
          temp_g=g[x][y]+1 #(g[0][3]=0)+1
          if temp_g<g[nx][ny]: #1<g[1][3]=yes as 1,3 is inf
            g[nx][ny]=temp_g #g[1][3]=1
            source[(nx,ny)]=(x,y,rasta) #source[(1,3)]:(0,3,'D')
            f=temp_g+manhattan((nx,ny),goal) #1+(man(1,3),(6,5)=3)==4
            heapq.heappush(pq,[f,nx,ny]) pq=[[4,1,3]]
  path=""
  x,y=goal
  if (x,y) not in source:
    return -1,-1
  while (x,y)!=start:
    px,py,rasta=source[(x,y)]
    path=rasta+path
    x,y=px,py
  return len(path),path
def read_inpt(input1):
  with open(input1,'r') as f:
    n,m=map(int,f.readline().split())
    start=tuple(map(int,f.readline().split()))
    goal=tuple(map(int,f.readline().split()))
    maze=[]
    for _ in range(n):
      maze.append(list(f.readline().strip()))
  return start,goal,maze
def write_outpt(output1,path_length,path):
  with open(output1,'w') as f:
    if path_length==-1:
      f.write("not found")
    else:
      f.write(f"{path_length}\n{path}")
def main(input1,output1):
  start,goal,maze=read_inpt(input1) #maze becomes a list of 7 lists
  path_length,path=astar(maze,start,goal)
  write_outpt(output1,path_length,path)
if __name__=='__main__':
  main('input1.txt','output1.txt')
```

```python
#2
with open("input2.txt", "w") as f:
    f.write("""5 5
    2 4
    1 0
    2 2
    3 2
    4 0
    5 1
    5 2
    2 3
    1 4
    4 5
    5 3

""")
def astar(graph,heuristics,start,goal):
  pq=[(heuristics[start],start)]
  g_cost={}
  for node in graph:
      g_cost[node] = float('inf')
  g_cost[start]=0
  source={start:None}
  while pq:
    f,curr=heapq.heappop(pq)
    if curr==goal:
      break
    else:
      for next in graph[curr]:
        temp_g=g_cost[curr]+1
        if temp_g<g_cost[next]:
          g_cost[next]=temp_g
          source[next]=curr
          heapq.heappush(pq,(g_cost[next]+heuristics[next],next))
  return g_cost.get(goal, float('inf'))
def read_graph(x):
  graph={}
  heuristics={}
  with open(x,'r') as f:
    n,m=map(int,f.readline().split())
    start,goal=f.readline().split()
    for i in range (n):
      vertex,h=f.readline().split()
      heuristics[vertex]=int(h)
      graph[vertex]=[]
    for j in range(m):
      u,v=f.readline().split()
      graph[u].append(v)
      graph[v].append(u)
  return graph,heuristics,start,goal
def main():
  inpt2='input2.txt'
  outpt2='output2.txt'
  graph,heuristics,start,goal=read_graph(inpt2)
  inadmissible=[]
  for node in graph:
    actual_cost=astar(graph,heuristics,node,goal)
    if heuristics[node] > actual_cost:
      inadmissible.append(node)
  with open(outpt2,'w') as f:
    if not inadmissible:
      f.write("1")
      print("1")
    else:
      f.write("0")
      print("0")
      f.write("Here nodes " + ", ".join(sorted(inadmissible)) + " are inadmissible.\n")
      print(f"Here nodes {', '.join(sorted(inadmissible))} are inadmissible.")

if __name__=='__main__':
  main()
```
